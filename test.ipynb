```python
import requests
import json
import os
from dotenv import load_dotenv

load_dotenv()


class SerperClient:
    def __init__(self, api_key=str):
        self.api_key = api_key or os.getenv("SERPER_API_KEY")
        self.base_url = "https://google.serper.dev"
        self.headers = {
            'X-API-KEY': self.api_key,
            'Content-Type': 'application/json'
        }

    def _make_request(self, endpoint, payload):
        response = requests.request("POST", self.base_url + endpoint, headers=self.headers, data=payload)
        data = response.read()
        return data.decode("utf-8")

    def search_query(self, query):
        payload = json.dumps({
            "q": query,
            "gl" :"in" # Location   # "gl": "in" for India
            })
        return self._make_request("/search", payload)

    def image_query(self, keywords):
        payload = json.dumps({
            "q": keywords,
            "gl": "in" # Location   # "gl": "in" for India
        })
        return self._make_request("/images", payload)
```

```python
import requests
from requests.exceptions import RequestException
from bs4 import BeautifulSoup
class Scraper:
    def __init__(self):
        pass

    def contentExtractor(self, response):
        soup = BeautifulSoup(response, 'html.parser')
        return  {
            "all_paragraphs" : [i.text.replace("\n", " ") for i in soup.find_all('p')],
            "all_headings" : [i.text for i in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])],
        }
        

    def get_website_content(self, link, timeout=10, headers=None):
        """
        Fetch content from a website URL
        
        Args:
            link (str): URL of the website to fetch
            timeout (int): Request timeout in seconds (default: 10)
            headers (dict): Optional custom headers to send with the request
        
        Returns:
            str: Website content as text
            
        Raises:
            RequestException: If any network-related error occurs
            ValueError: If the response contains an HTTP error
        """
        default_headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
        }
        
        headers = headers or default_headers

        try:
            response = requests.get(
                link,
                headers=headers,
                timeout=timeout,
                allow_redirects=True
            )
            response.raise_for_status()  # Raise exception for HTTP errors
            
            return self.contentExtractor(response.content)
            
        except RequestException as e:
            raise RequestException(f"Failed to fetch {link}: {str(e)}") from e
        except Exception as e:
            raise RuntimeError(f"Unexpected error occurred: {str(e)}") from e    
            
```
